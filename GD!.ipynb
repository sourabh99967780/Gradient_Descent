```python
import numpy as np
```

```python
def step_gradient(data,learning_rate,m):
    m_slope = np.zeros(len(m))
    row = data.shape[0]
    col = m.shape[0]
    x = data[:,0:data.shape[1]-1]
    y = data[:,data.shape[1] - 1:]
    for i in range(col):
        for j in range(row):
            xj = x[j,:]
            yj = y[j,:]
            m_slope[i] += (-2/row)*(yj - m.dot(xj))*(xj[i])
        new_m = m - learning_rate * m_slope
    return new_m
```

```python
def cost(data,m):
    total_cost = np.zeros(len(m))
    row = data.shape[0]
    col = m.shape[0]
    x = data[:,0:data.shape[1]-1]
    y = data[:,data.shape[1] - 1:]
    for i in range(col):
        for j in range(row):
            xj = x[j,:]
            yj = y[j,:]
            total_cost[i] += (1/row)*((yj - m.dot(xj))**2)
            
    return total_cost
```

```python
def gd(data,learning_rate,num_iterations):
    m = np.zeros(data.shape[1]-1)
    for i in range(num_iterations):
        m = step_gradient(data,learning_rate,m)
        print(i,"Cost: ",cost(data,m))
    return m
```

```python
def run():
    data = np.loadtxt("training_boston_x_y_train.csv",delimiter=",")
    row = data.shape[0]
    col = data.shape[1]
    x = data[:,0:data.shape[1]-1]
    y = data[:,data.shape[1]-1:]
    print(x.shape)
    print(y.shape)
    z = np.ones((row,1))
    new_d = np.hstack((x,z))
    new_d = np.hstack((new_d,y))
    learning_rate = 0.1
    num_iterations = 100
    m = gd(new_d,learning_rate,num_iterations)
    
    # predicting on the test data
    test_data = np.loadtxt("test_boston_x_test.csv",delimiter=",")
    z = np.ones((test_data.shape[0],1))
    new_xtest = np.hstack((test_data,z))
    y_pred = new_xtest.dot(m)
    # np.savetxt("Boston_dataset_prediction.csv",y_pred,fmt="%.7f")
    print(y_pred)
```

```python
run()
```
